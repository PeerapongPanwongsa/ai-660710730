##การทำให้ข้อมูลเป็นมาตรฐาน

```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'Science', 'วิทย์', 'คณะวิทยาศาสตร์', 'Sci']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'Arts', 'ศิลป์', 'Liberal Arts']:
        return 'ศิลปศาสตร์'
    elif name in ['วิศวะ', 'Engineering', 'วิศวฯ', 'วิศวกรรมศาสตร์']:
        return 'วิศวกรรมศาสตร์'
    else:
        return name
```

```python
df['คณะ'] = df['คณะ'].apply(standardize_faculty)
```

```python
df['คณะ']
```

```python
def standardize_faculty(sex):
    if sex in ['ชาย', 'Male', 'ช', 'M']:
        return 'ชาย'
    elif sex in ['หญิง', 'Female', 'ญ', 'F']:
        return 'หญิง'
    else:
        return sex
```

```python
df['เพศ'] = df['เพศ'].apply(standardize_faculty)
```

##การจัดการค่าผิดปกติ Q1,Q3,IQR

```python
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
```

```python
Q1, Q3
```

```python
IQR = Q3 - Q1
```

```python
IQR
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
lower_bound, upper_bound
```

```python
df_no_outliers = df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)]
```

```python
df['อายุ'] >= lower_bound
```

```python
df['อายุ'] <= upper_bound
```

```python
print(f"\nขอบเขตล่าง - {lower_bound}")
print(f"ขอบเขตบน - {upper_bound}")
print(f"\nข้อมูลหลังกรองค่าผิดปกติ")
print(df_no_outliers)
print(f"\nค่าเฉลี่ยอายุ (หลังกรองค่าผิดปกติ) - {df_no_outliers['อายุ'].mean():.2f}")
```

##การสร้าง Feature จากวันเวลา : หมายเหตุผมทำไม่เป็น เพราะในข้อมูลตัวอย่างมีเวลาให้ แต่อันนี้ไม่มี ผมเลยลองให้ AI ช่วย และทำความเข้าใจ

```python
df['วันที่ลงทะเบียน'] = pd.to_datetime(df['วันที่ลงทะเบียน'], errors='coerce')
```

```python
df['ชั่วโมง'] = df['วันที่ลงทะเบียน'].dt.hour  # ถ้าไม่มีเวลาจะเป็น 0
df['วันในสัปดาห์'] = df['วันที่ลงทะเบียน'].dt.day_name()
df['เดือน'] = df['วันที่ลงทะเบียน'].dt.month
df['ปี'] = df['วันที่ลงทะเบียน'].dt.year
```

```python
df['ช่วงเวลา'] = pd.cut(
    df['ชั่วโมง'],
    bins=[0, 10, 14, 18, 24],
    labels=['เช้า', 'กลางวัน', 'บ่าย', 'เย็น'],
    right=False
)
```
